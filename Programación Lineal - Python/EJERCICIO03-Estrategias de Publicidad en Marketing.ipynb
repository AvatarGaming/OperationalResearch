## Selecci√≥n de Medios

Los modelos de programaci√≥n lineal se han utilizado en el campo de la publicidad como una ayuda para la toma de decisiones en la selecci√≥n de una mezcla de medios efectiva. A veces, la t√©cnica se emplea para asignar un presupuesto fijo o limitado entre varios medios, que pueden incluir comerciales de radio o televisi√≥n, anuncios en peri√≥dicos, env√≠os por correo directo, redes sociales, y otros. En otras aplicaciones, el objetivo es la maximizaci√≥n de la exposici√≥n de la audiencia. Las restricciones en la mezcla de medios permitida pueden surgir debido a requisitos contractuales, disponibilidad limitada de medios o pol√≠ticas de la empresa. A continuaci√≥n, se presenta un ejemplo.

El Club de Juegos "Win Big" promueve viajes de juego desde una gran ciudad del medio oeste a casinos en las Bahamas. El club ha presupuestado hasta $8,000 por semana para publicidad local. El dinero se debe asignar entre cuatro medios promocionales: spots de televisi√≥n, anuncios en peri√≥dicos, y dos tipos de anuncios de radio. El objetivo de Win Big es llegar a la mayor cantidad posible de audiencia potencial de alto valor a trav√©s de los diversos medios. La siguiente tabla presenta el n√∫mero de potenciales jugadores alcanzados mediante el uso de un anuncio en cada uno de los cuatro medios. Tambi√©n proporciona el costo por anuncio colocado y el n√∫mero m√°ximo de anuncios que se pueden comprar por semana.

| MEDIO | AUDIENCIA ALCANZADA POR ANUNCIO | COSTO POR ANUNCIO ($) | M√ÅXIMO DE ANUNCIOS POR SEMANA |
|-------|----------------------------------|------------------------|-------------------------------|
| Spot de TV (1 minuto) | 5,000 | 800 | 12 |
| Peri√≥dico diario (anuncio de p√°gina completa) | 8,500 | 925 | 5 |
| Spot de radio (30 segundos, hora pico) | 2,400 | 290 | 25 |
| Spot de radio (1 minuto, tarde) | 2,800 | 380 | 20 |

Los acuerdos contractuales de Win Big requieren que se coloquen al menos cinco spots de radio cada semana. Para asegurar una campa√±a promocional de amplio alcance, la gerencia tambi√©n insiste en que no se gaste m√°s de $1,800 en publicidad de radio cada semana.

Al formular esto como un problema de programaci√≥n lineal, el primer paso es comprender completamente el problema. A veces, hacer preguntas de "¬øqu√© pasar√≠a s√≠?" ayudar√° a entender la situaci√≥n. En este ejemplo, ¬øqu√© pasar√≠a si se usaran exactamente cinco anuncios de cada tipo? ¬øCu√°nto costar√≠an los anuncios? ¬øA cu√°ntas personas alcanzar√≠an? Sin duda, el uso de una hoja de c√°lculo para los c√°lculos puede ayudar con esto, ya que se pueden escribir f√≥rmulas para calcular el costo y el n√∫mero de personas alcanzadas. Una vez que se entiende la situaci√≥n, se establecen el objetivo y las restricciones:

### Objetivo:
Maximizar el n√∫mero de personas (audiencia) alcanzadas

### Restricciones:
a) No se pueden usar m√°s de 12 anuncios de TV.  
b) No se pueden usar m√°s de 5 anuncios en peri√≥dicos.  
c) No se pueden usar m√°s de 25 anuncios de radio de 30 segundos.  
d) No se pueden usar m√°s de 20 anuncios de radio de 1 minuto.  
e) El monto total gastado no puede exceder los $8,000.  
f) El n√∫mero total de anuncios de radio debe ser al menos 5.  
g) El monto total gastado en anuncios de radio no debe exceder los $1,800.  

### Variables de decisi√≥n:
A continuaci√≥n, defina las variables de decisi√≥n. Las decisiones que se est√°n tomando aqu√≠ involucran el n√∫mero de anuncios de cada tipo a utilizar. Una vez que se conocen, se pueden usar para calcular el monto gastado y el n√∫mero de personas alcanzadas.

Sea:
- $X_1$ = n√∫mero de spots de TV de 1 minuto tomados cada semana
- $X_2$ = n√∫mero de anuncios de p√°gina completa en el peri√≥dico diario tomados cada semana
- $X_3$ = n√∫mero de spots de radio de 30 segundos en hora pico tomados cada semana
- $X_4$ = n√∫mero de spots de radio de 1 minuto en la tarde tomados cada semana

### Ecuaciones
A continuaci√≥n, usando estas variables, escriba la expresi√≥n matem√°tica para el objetivo y las restricciones que se identificaron. Las restricciones de no negatividad tambi√©n se explicitan.

**Objetivo:**  
Maximizar la cobertura de la audiencia = $5000X_1 + 8500X_2 + 2400X_3 + 2800X_4$

**(Restricciones) Sujeto a:**  
a) $X_1 \leq 12$ (m√°ximo de spots de TV por semana).  
b) $X_2 \leq 5$ (m√°ximo de anuncios en el peri√≥dico por semana).  
c) $X_3 \leq 25$ (m√°ximo de spots de radio de 30 segundos por semana).  
d) $X_4 \leq 20$ (m√°ximo de spots de radio de 1 minuto por semana).  
e) $800X_1 + 925X_2 + 290X_3 + 380X_4 \leq 8000$ (presupuesto total no puede exceder los $8000).  
f) $X_3 + X_4 \geq 5$ (al menos 5 anuncios de radio por semana).  
g) $290X_3 + 380X_4 \leq 1800$ (presupuesto en anuncios de radio no puede exceder $1800).  

Estas ecuaciones describen el problema de programaci√≥n lineal para maximizar la audiencia alcanzada dentro de las restricciones presupuestarias y de cantidad de anuncios.

# üìä Definici√≥n de Datos del Problema

## üéØ Funci√≥n Principal

El **PASO 2** del script transforma los datos del ejercicio "Win Big" en estructuras de Python para su manipulaci√≥n algor√≠tmica:

| üîÑ Estructura | üìã Descripci√≥n | üìù Ejemplo |
|---------------|---------------|-----------|
| üì∫ `medios` | Lista de canales publicitarios | `['TV', 'Peri√≥dico', 'Radio_pico', 'Radio_tarde']` |
| üë• `audiencia` | Diccionario de alcance por medio | `'TV': 5000` |
| üí∞ `costo` | Diccionario de precio por anuncio | `'Peri√≥dico': 925` |
| üî¢ `max_anuncios` | L√≠mites semanales por medio | `'Radio_pico': 25` |

Tambi√©n define las constantes:
* üíµ Presupuesto total: $8,000
* üìª Presupuesto radio: $1,800 
* üì£ M√≠nimo spots radio: 5

Estas estructuras preparan los datos para la funci√≥n objetivo:
$$ \text{Maximizar } 5000X_1 + 8500X_2 + 2400X_3 + 2800X_4 $$

````python
# PASO 2: DEFINIR LOS DATOS DEL PROBLEMA
# ---------------------------------------
# Definimos los par√°metros del problema en estructuras de datos de Python.
# Esto facilita la modificaci√≥n de los datos si es necesario.

# Tipos de medios publicitarios
medios = ['TV', 'Peri√≥dico', 'Radio_pico', 'Radio_tarde']

# Datos de cada medio: audiencia alcanzada, costo y m√°ximo de anuncios
audiencia = {
    'TV': 5000,         # Personas alcanzadas por un spot de TV
    'Peri√≥dico': 8500,  # Personas alcanzadas por un anuncio de peri√≥dico
    'Radio_pico': 2400, # Personas alcanzadas por un spot de radio en hora pico
    'Radio_tarde': 2800 # Personas alcanzadas por un spot de radio en la tarde
}

costo = {
    'TV': 800,         # Costo por spot de TV ($)
    'Peri√≥dico': 925,  # Costo por anuncio de peri√≥dico ($)
    'Radio_pico': 290, # Costo por spot de radio en hora pico ($)
    'Radio_tarde': 380 # Costo por spot de radio en la tarde ($)
}

max_anuncios = {
    'TV': 12,         # M√°ximo de spots de TV por semana
    'Peri√≥dico': 5,   # M√°ximo de anuncios de peri√≥dico por semana
    'Radio_pico': 25, # M√°ximo de spots de radio en hora pico por semana
    'Radio_tarde': 20 # M√°ximo de spots de radio en la tarde por semana
}

# Presupuesto total disponible por semana
presupuesto_total = 8000  # en d√≥lares

# Presupuesto m√°ximo para anuncios de radio
presupuesto_radio = 1800  # en d√≥lares

# M√≠nimo de spots de radio requeridos
min_spots_radio = 5

# Funci√≥n para visualizar la informaci√≥n de las variables
def visualizar_info(variable, nombre):
    print(f"üìä Informaci√≥n de la variable: {nombre}")
    print(f"üîç Tipo de variable: {type(variable)}")
    print(f"üìè Longitud/Estructura: {len(variable) if hasattr(variable, '__len__') else 'N/A'}")
    print(f"üìÑ Contenido:")
    print(variable)
    print("="*25)

# Visualizar la informaci√≥n de cada variable
visualizar_info(medios, "medios")
visualizar_info(audiencia, "audiencia")
visualizar_info(costo, "costo")
visualizar_info(max_anuncios, "max_anuncios")
visualizar_info(presupuesto_total, "presupuesto_total")
visualizar_info(presupuesto_radio, "presupuesto_radio")
visualizar_info(min_spots_radio, "min_spots_radio")
````

# üß© Explicaci√≥n del PASO 3: Modelo de Programaci√≥n Lineal

## üéØ Funci√≥n Principal

El PASO 3 establece la **estructura fundamental** del problema de optimizaci√≥n para "Win Big":

| üõ†Ô∏è Componente | üìù Descripci√≥n | üßÆ Implementaci√≥n |
|---------------|----------------|-------------------|
| üìä **Tipo de problema** | Maximizaci√≥n de audiencia | `pulp.LpProblem(sense=pulp.LpMaximize)` |
| üî¢ **Variables de decisi√≥n** | Cantidad de anuncios ($X_1, X_2, X_3, X_4$) | `x = pulp.LpVariable.dicts("anuncios", medios, lowBound=0, cat=pulp.LpInteger)` |

Este paso traduce la formulaci√≥n matem√°tica en c√≥digo ejecutable, creando el contenedor para la funci√≥n objetivo $Z = 5000X_1 + 8500X_2 + 2400X_3 + 2800X_4$ y las restricciones que ser√°n definidas posteriormente.

üí° Las variables se definen como enteras y no negativas, reflejando que no podemos comprar fracciones de anuncios ni cantidades negativas.

````python
# PASO 3: CREAR EL MODELO DE PROGRAMACI√ìN LINEAL
# ----------------------------------------------
# Aqu√≠ creamos el modelo de PuLP y definimos las variables de decisi√≥n,
# la funci√≥n objetivo y las restricciones.

# Crear un problema de maximizaci√≥n
# En programaci√≥n lineal, podemos maximizar o minimizar una funci√≥n objetivo
problema = pulp.LpProblem(name="Maximizacion_Audiencia", sense=pulp.LpMaximize)

# Definir las variables de decisi√≥n
# LpVariable.dicts crea un diccionario de variables, una para cada medio
# Las variables son no negativas (lowBound=0) y enteras (cat='Integer')
x = pulp.LpVariable.dicts("anuncios", medios, lowBound=0, cat=pulp.LpInteger)

visualizar_info(problema, "problema")
visualizar_info(x, "x")
````

# üéØ PASO 4: Definici√≥n de la Funci√≥n Objetivo

## üìä Prop√≥sito
El PASO 4 establece la **funci√≥n objetivo** para maximizar el alcance total de audiencia del Club de Juegos "Win Big".

## üßÆ F√≥rmula Matem√°tica
$$ \text{Maximizar} \sum_{m \in \text{medios}} \text{audiencia}[m] \cdot x[m] = 5000 X_1 + 8500 X_2 + 2400 X_3 + 2800 X_4 $$

## üíª Implementaci√≥n
```python
problema += pulp.lpSum([audiencia[m] * x[m] for m in medios]), "Audiencia_Total"
```

## üìù Explicaci√≥n T√©cnica

| üîç Componente | ‚ú® Funci√≥n |
|--------------|-----------|
| `pulp.lpSum()` | üî¢ Suma los productos de audiencia alcanzada por anuncios |
| `[audiencia[m] * x[m] for m in medios]` | üîÑ List comprehension que multiplica cada coeficiente de audiencia por su variable |
| `"Audiencia_Total"` | üè∑Ô∏è Etiqueta que identifica la funci√≥n objetivo |

üöÄ Esta l√≠nea traduce el objetivo matem√°tico a c√≥digo ejecutable, permitiendo al solucionador buscar valores √≥ptimos para las variables de decisi√≥n.

````python
# PASO 4: DEFINIR LA FUNCI√ìN OBJETIVO
# -----------------------------------
# La funci√≥n objetivo es maximizar la audiencia alcanzada
# Expresi√≥n: 5000*X_1 + 8500*X_2 + 2400*X_3 + 2800*X_4

# Crear la funci√≥n objetivo sumando la audiencia alcanzada por cada medio
problema += pulp.lpSum([audiencia[m] * x[m] for m in medios]), "Audiencia_Total"
visualizar_info(problema, "problema")
````

# üìä PASO 5: IMPLEMENTACI√ìN DE RESTRICCIONES

## üîí Funci√≥n en el Modelo de Programaci√≥n Lineal

El **PASO 5** traduce matem√°ticamente las limitaciones del problema "Win Big" al lenguaje de programaci√≥n:

| üî¢ | ‚ö†Ô∏è Restricci√≥n | üßÆ F√≥rmula Matem√°tica | üíª Implementaci√≥n Python |
|-----|---------------|----------------------|--------------------------|
| a | üì∫ M√°ximo spots TV | $X_1 \leq 12$ | `x['TV'] <= max_anuncios['TV']` |
| b | üì∞ M√°ximo anuncios peri√≥dico | $X_2 \leq 5$ | `x['Peri√≥dico'] <= max_anuncios['Peri√≥dico']` |
| c | üìª M√°ximo spots radio (pico) | $X_3 \leq 25$ | `x['Radio_pico'] <= max_anuncios['Radio_pico']` |
| d | üéôÔ∏è M√°ximo spots radio (tarde) | $X_4 \leq 20$ | `x['Radio_tarde'] <= max_anuncios['Radio_tarde']` |
| e | üí∞ Presupuesto total | $800X_1 + 925X_2 + 290X_3 + 380X_4 \leq 8000$ | `pulp.lpSum([costo[m] * x[m] for m in medios]) <= presupuesto_total` |
| f | üîÑ M√≠nimo spots radio | $X_3 + X_4 \geq 5$ | `x['Radio_pico'] + x['Radio_tarde'] >= min_spots_radio` |
| g | üí∏ L√≠mite gasto radio | $290X_3 + 380X_4 \leq 1800$ | `costo['Radio_pico']*x['Radio_pico'] + costo['Radio_tarde']*x['Radio_tarde'] <= presupuesto_radio` |

Este c√≥digo üîç define el **espacio factible** donde el algoritmo buscar√° la soluci√≥n √≥ptima para maximizar el alcance de audiencia.

````python
# PASO 5: A√ëADIR LAS RESTRICCIONES
# --------------------------------
# A√±adimos todas las restricciones del problema al modelo

# Restricci√≥n (a): M√°ximo de spots de TV por semana
problema += x['TV'] <= max_anuncios['TV'], "Max_TV"

# Restricci√≥n (b): M√°ximo de anuncios de peri√≥dico por semana
problema += x['Peri√≥dico'] <= max_anuncios['Peri√≥dico'], "Max_Periodico"

# Restricci√≥n (c): M√°ximo de spots de radio en hora pico por semana
problema += x['Radio_pico'] <= max_anuncios['Radio_pico'], "Max_Radio_Pico"

# Restricci√≥n (d): M√°ximo de spots de radio en la tarde por semana
problema += x['Radio_tarde'] <= max_anuncios['Radio_tarde'], "Max_Radio_Tarde"

# Restricci√≥n (e): Presupuesto total no puede exceder $8000
problema += pulp.lpSum([costo[m] * x[m] for m in medios]) <= presupuesto_total, "Presupuesto_Total"

# Restricci√≥n (f): Al menos 5 anuncios de radio por semana
problema += x['Radio_pico'] + x['Radio_tarde'] >= min_spots_radio, "Min_Anuncios_Radio"

# Restricci√≥n (g): Presupuesto en anuncios de radio no puede exceder $1800
problema += costo['Radio_pico']*x['Radio_pico'] + costo['Radio_tarde']*x['Radio_tarde'] <= presupuesto_radio, "Presupuesto_Radio"

visualizar_info(problema, "problema")
````

# üöÄ PASO 6: RESOLVER EL MODELO 

## üéØ Funci√≥n Principal

El **PASO 6** es el punto crucial donde el algoritmo de optimizaci√≥n encuentra la distribuci√≥n √≥ptima de anuncios para el Club "Win Big". 

| üìä Componente | üîç Descripci√≥n |
|--------------|---------------|
| `problema.solve()` | üßÆ Ejecuta el algoritmo simplex para encontrar valores de $X_1$, $X_2$, $X_3$ y $X_4$ que maximicen $Z = 5000X_1 + 8500X_2 + 2400X_3 + 2800X_4$ |
| `pulp.LpStatus[problema.status]` | ‚úÖ Verifica si se encontr√≥ una soluci√≥n √≥ptima o si hubo alg√∫n error |

## üí° Resultado

Cuando finaliza, el modelo determina **exactamente cu√°ntos anuncios** de cada medio maximizan la audiencia total respetando:

$$\begin{cases}
X_1 \leq 12 \\
X_2 \leq 5 \\
X_3 \leq 25 \\
X_4 \leq 20 \\
800X_1 + 925X_2 + 290X_3 + 380X_4 \leq 8000 \\
X_3 + X_4 \geq 5 \\
290X_3 + 380X_4 \leq 1800
\end{cases}$$

üîÑ Este paso transforma la teor√≠a en acci√≥n, materializando la soluci√≥n del problema de programaci√≥n lineal.

````python
# PASO 6: RESOLVER EL MODELO
# --------------------------
# Utilizamos el solucionador para encontrar la soluci√≥n √≥ptima

# Resolver el problema
problema.solve()

# Verificar el estado de la soluci√≥n
# pulp.LpStatus[problema.status] devolver√° 'Optimal' si se encuentra una soluci√≥n √≥ptima
print(f"\nEstado de la soluci√≥n: {pulp.LpStatus[problema.status]}")

````

# üîç Funci√≥n del PASO 7: Mostrar Resultados

## üìä Descripci√≥n General

El **PASO 7** procesa y presenta la soluci√≥n √≥ptima del problema de programaci√≥n lineal para el Club de Juegos "Win Big".

## üßÆ Operaciones Principales

| Operaci√≥n | Descripci√≥n | F√≥rmula |
|-----------|-------------|---------|
| üìà Extracci√≥n de valores | Obtiene valores √≥ptimos de variables | `valor = int(x[m].value())` |
| üî¢ C√°lculo de audiencia | Suma la audiencia alcanzada | $\sum_{m \in medios} audiencia_m \times X_m$ |
| üí∞ C√°lculo de costo | Suma el costo total | $\sum_{m \in medios} costo_m \times X_m$ |
| ‚öñÔ∏è C√°lculo de eficiencia | Determina costo por persona | $\frac{costo\_total}{audiencia\_total}$ |
| ‚úÖ Verificaci√≥n | Comprueba restricciones especiales | $X_3 + X_4 \geq 5$ |

## üéØ Importancia

Este bloque es **crucial** ya que transforma los resultados matem√°ticos en informaci√≥n accionable para la toma de decisiones publicitarias, verificando el cumplimiento de todas las restricciones contractuales del modelo original.

$$Z = \sum_{i=1}^{n} c_i x_i$$

$$\Delta Z \approx y_j \cdot \Delta b_j$$
