# 📚 Problema de Horarios de Exámenes Universitarios

## 🎯 ¿Qué estamos tratando de resolver?

Imagina que eres el encargado de programar los exámenes en una universidad. Tienes 14 exámenes diferentes que debes colocar en días distintos. Tu objetivo es usar la menor cantidad de días posible para programar todos los exámenes.

## 📋 Los exámenes que tenemos

La universidad tiene 3 materias principales:
- **🎨 Arte**: 5 exámenes (2 obligatorios y 3 opcionales)
- **🧬 Biología**: 4 exámenes (2 obligatorios y 2 opcionales)
- **⚗️ Química**: 5 exámenes (2 obligatorios y 3 opcionales)

### Identificación de los exámenes

| 🎨 Arte | 🧬 Biología | ⚗️ Química |
|:-------:|:----------:|:----------:|
| **Obligatorios** | **Obligatorios** | **Obligatorios** |
| A1 | B1 | Q1 |
| A2 | B2 | Q2 |
| **Opcionales** | **Opcionales** | **Opcionales** |
| A3 | B3 | Q3 |
| A4 | B4 | Q4 |
| A5 | - | Q5 |

## ⚠️ El problema principal

No podemos programar ciertos exámenes en el mismo día porque algunos estudiantes deben presentar varios exámenes. Un estudiante no puede estar en dos lugares al mismo tiempo.

## 📏 Las reglas que debemos seguir

1. **Regla 1** ⛔: Ningún examen de Arte puede estar con otro examen de Arte en el mismo día (porque hay estudiantes que toman varios cursos de Arte).

2. **Regla 2** ⛔: Los exámenes obligatorios de Arte no pueden estar el mismo día que cualquier examen de Biología (porque hay estudiantes de Biología que también toman clases obligatorias de Arte).

3. **Regla 3** ⛔: Los exámenes opcionales de Química no pueden estar el mismo día que cualquier examen de Biología (porque hay estudiantes de Biología que también toman clases opcionales de Química).

4. **Regla 4** ⛔: Los exámenes obligatorios de Química no pueden estar el mismo día que cualquier examen opcional de Biología (porque hay estudiantes que toman ambos).

## 🧮 ¿Cómo resolvemos este problema?

Para resolver este problema usamos algo llamado "variables de decisión":

- Creamos una variable X para cada combinación de examen y día
- Por ejemplo, $X_{(A1,Día1)} = 1$ significa "El examen A1 se programa en el Día 1"
- $X_{(A1,Día1)} = 0$ significa "El examen A1 NO se programa en el Día 1"

También creamos otra variable Y para cada día:
- $Y_{(Día1)} = 1$ significa "Usamos el Día 1 para algún examen"
- $Y_{(Día1)} = 0$ significa "No usamos el Día 1 para ningún examen"

## 🎯 Nuestro objetivo

Queremos minimizar el número total de días usados. Matemáticamente, esto es:

$$\text{Minimizar: } \sum_{t=1}^{14} Y_t = Y_1 + Y_2 + Y_3 + ... + Y_{14}$$

## 🚧 Restricciones en palabras simples

1. **Cada examen debe programarse exactamente una vez** ✅  
   Si tenemos el examen A1, debe aparecer en exactamente un día, ni más ni menos.
   
   $$\sum_{t=1}^{14} X_{(A1,t)} = 1$$

2. **Si usamos un día, debemos marcar ese día como "usado"** 🔄  
   Si programamos cualquier examen en el Día 3, entonces $Y_{(Día3)}$ debe ser 1.
   
   $$\text{Si } X_{(cualquier_examen,Día3)} = 1, \text{ entonces } Y_{(Día3)} = 1$$

3. **No podemos poner exámenes con conflictos en el mismo día** ❌  
   Por ejemplo, no podemos poner A1 y A2 en el mismo día (por la Regla 1).
   
   $$X_{(A1,Día1)} + X_{(A2,Día1)} \leq 1 \text{ para cualquier día}$$

En este ejemplo, necesitaríamos 6 días para programar todos los exámenes respetando todas las reglas. Pero quizás existe una solución que usa menos días, y ese es el desafío de la programación lineal entera.

## 📊 ¿Qué es la programación lineal entera?

Es una técnica matemática para encontrar la mejor solución posible a problemas como este. En lugar de probar todas las combinaciones posibles (que serían muchísimas), usamos algoritmos especiales que encuentran la solución óptima rápidamente.

### Formulación matemática generalizada:

$$
\begin{align*}
\text{Minimizar: } & \sum_{t=1}^{T} Y_t \\
\text{Sujeto a: } & \sum_{t=1}^{T} X_{m,t} = 1 \quad \forall m \in \text{Exámenes} \\
& X_{m,t} \leq Y_t \quad \forall m \in \text{Exámenes}, \forall t \in \text{Días} \\
& \sum_{m \in \text{Grupo de conflicto}} X_{m,t} \leq 1 \quad \forall t \in \text{Días} \\
& X_{m,t}, Y_t \in \{0,1\} \quad \forall m \in \text{Exámenes}, \forall t \in \text{Días}
\end{align*}
$$

Esta es la técnica que utilizan 🏫 universidades reales, ✈️ aerolíneas, 🏥 hospitales y muchas empresas para resolver problemas complejos de programación y asignación de recursos.

## 🏆 Beneficios de esta solución

- ⏱️ Ahorra tiempo al usar el mínimo número de días para los exámenes
- 🧠 Evita conflictos para los estudiantes que toman múltiples asignaturas
- 📝 Asegura que cada examen se programe exactamente una vez
- 🔄 Permite ajustes si cambian las restricciones o se añaden más exámenes

# 📊 Explicación del Código Python para el Problema de Exámenes

Esta función `get_variables()` crea las variables de decisión necesarias para resolver nuestro problema de horarios.

## 🔢 Variables Binarias

La función genera dos tipos de variables binarias:

- **`x[m,t]`**: Indica si el módulo `m` está programado en el tiempo `t` (1 = sí, 0 = no).
- **`y[t]`**: Indica si se usa el tiempo `t` (1 = sí, 0 = no).

## 🧩 Uso de la Biblioteca `pulp`

La función utiliza la biblioteca `pulp`, una herramienta poderosa para resolver problemas de programación lineal.

## 📝 Estructura de las Variables

| Variable | Significado               | Valores |
|:--------:|:-------------------------:|:-------:|
| `x[m,t]` | ¿Examen `m` en tiempo `t`? | 0 ó 1   |
| `y[t]`   | ¿Se usa el tiempo `t`?     | 0 ó 1   |

Estas variables corresponden exactamente a las $X_{m,t}$ y $Y_t$ de nuestra formulación matemática. ✅

````python
# 📚 Importamos la biblioteca PuLP para resolver problemas de programación lineal
import pulp

# 🧮 Definición de la función para crear variables de decisión
def get_variables(modules, times):
    """🔢 Returns the binary variables for a given timetabling
    problem.
    
    Esta función genera las variables de decisión binarias necesarias
    para modelar nuestro problema de programación de exámenes.
    
    Args:
        modules: The complete collection of modules to be
                timetabled. 
                # 📝 Lista de todos los exámenes que necesitamos programar (A1, A2, B1, etc.)
        
        times: The collection of available time slots.
               # ⏰ Lista de franjas horarias disponibles (día1, día2, etc.)
    
    Returns:
        A tuple containing the decision variables x and y.
        # 🔄 Devuelve las variables binarias X y Y del modelo
    """
    
    # 📊 Forma dimensional para la variable X (matriz módulos x tiempos)
    # Cada X[m,t] representa si el módulo m está programado en el tiempo t
    xshape = (modules, times)
    
    # ✅ Creamos variables X[m,t] de tipo binario (0 o 1)
    # X[m,t] = 1 significa "el módulo m está programado en el tiempo t"
    # X[m,t] = 0 significa "el módulo m NO está programado en el tiempo t"
    # En programación lineal entera, estas son nuestras variables de decisión principales
    x = pulp.LpVariable.dicts("X", xshape, cat=pulp.LpBinary)
    
    # 🗓️ Creamos variables Y[t] de tipo binario (0 o 1)
    # Y[t] = 1 significa "la franja horaria t está siendo utilizada"
    # Y[t] = 0 significa "la franja horaria t NO está siendo utilizada"
    # En nuestro modelo, intentaremos minimizar la suma de estas variables
    y = pulp.LpVariable.dicts("Y", times, cat=pulp.LpBinary)
    
    # 🔄 Devolvemos ambas variables para usarlas en el modelo de optimización
    return x, y
````

# 🧩 Inicialización de Datos para el Problema de Exámenes

Esta celda define los grupos de módulos y crea las variables del problema.

## 🎨 Conjuntos de Módulos

Se definen los siguientes conjuntos de módulos:

| Variable | Significado           | Elementos       |
|:--------:|:---------------------:|:---------------:|
| `Ac`     | Arte obligatorio      | `[0, 1]`        |
| `Ao`     | Arte opcional         | `[2, 3, 4]`     |
| `Bc`     | Biología obligatoria  | `[5, 6]`        |
| `Bo`     | Biología opcional     | `[7, 8]`        |
| `Cc`     | Química obligatoria   | `[9, 10]`       |
| `Co`     | Química opcional      | `[11, 12, 13]`  |

## 📋 Lista Completa de Módulos

Se crea una lista combinando todos los grupos de módulos definidos anteriormente.

## ⏰ Franjas Horarias

Se generan **14 franjas horarias potenciales** utilizando `times = range(14)`.

## 🔢 Creación de Variables

Se llama a la función `get_variables()` para crear las variables binarias:
- **$X_{m,t}$**: Indica si el módulo `m` está programado en el tiempo `t`.
- **$Y_t$**: Indica si se usa el tiempo `t`.

````python
# 🎨 Definimos los módulos de Arte (A) obligatorios (c = core)
Ac = [0, 1]  # Estos corresponden a los exámenes A1 y A2 en nuestro problema
# 🎭 Definimos los módulos de Arte opcionales (o = optional)
Ao = [2, 3, 4]  # Estos corresponden a los exámenes A3, A4 y A5
# 🧬 Definimos los módulos de Biología (B) obligatorios
Bc = [5, 6]  # Estos corresponden a los exámenes B1 y B2
# 🔬 Definimos los módulos de Biología opcionales
Bo = [7, 8]  # Estos corresponden a los exámenes B3 y B4
# ⚗️ Definimos los módulos de Química (C) obligatorios
Cc = [9, 10]  # Estos corresponden a los exámenes Q1 y Q2
# 🧪 Definimos los módulos de Química opcionales
Co = [11, 12, 13]  # Estos corresponden a los exámenes Q3, Q4 y Q5

# Imprimir información sobre los módulos definidos
print("📋 INFORMACIÓN DE MÓDULOS:")
print(f"🎨 Módulos de Arte obligatorios (Ac): {Ac}")
print(f"🎭 Módulos de Arte opcionales (Ao): {Ao}")
print(f"🧬 Módulos de Biología obligatorios (Bc): {Bc}")
print(f"🔬 Módulos de Biología opcionales (Bo): {Bo}")
print(f"⚗️ Módulos de Química obligatorios (Cc): {Cc}")
print(f"🧪 Módulos de Química opcionales (Co): {Co}")

# 📚 Creamos la lista completa de todos los módulos 
# En Python, el operador + con listas concatena los elementos
modules = Ac + Ao + Bc + Bo + Cc + Co  # Lista completa de los 14 módulos
print(f"\n📚 Lista completa de módulos: {modules}")
print(f"   Total de módulos: {len(modules)}")

# ⏰ Creamos las posibles franjas horarias (en el peor caso, necesitaríamos 14 días)
# La función range(14) crea una secuencia de números del 0 al 13
times = range(14)  # Representa los posibles días para programar exámenes
print(f"\n⏰ Franjas horarias disponibles: {list(times)}")
print(f"   Total de franjas horarias: {len(list(times))}")

# 🧮 Generamos las variables binarias para nuestro modelo de programación lineal entera
# x[m,t] = 1 si el módulo m está programado en el tiempo t, 0 en caso contrario
# y[t] = 1 si se usa la franja t, 0 en caso contrario
print("\n🧮 Generando variables de decisión...")
x, y = get_variables(modules=modules, times=times)  # Usamos la función que definimos antes

# Imprimir información sobre las variables generadas
print("\n📊 RESUMEN DE VARIABLES:")
print(f"   • Variables X: {len(modules) * len(list(times))} variables binarias")
print(f"   • Variables Y: {len(list(times))} variables binarias")
print("\n📝 SIGNIFICADO:")
print("   • X[m,t] = 1 si el módulo m está programado en el tiempo t")
print("   • Y[t] = 1 si se utiliza la franja horaria t")

# Mostrar distribución por categoría
print("\n🔍 DISTRIBUCIÓN DE MÓDULOS:")
print(f"   • Módulos obligatorios: {len(Ac) + len(Bc) + len(Cc)} ({len(Ac)} Arte, {len(Bc)} Biología, {len(Cc)} Química)")
print(f"   • Módulos opcionales: {len(Ao) + len(Bo) + len(Co)} ({len(Ao)} Arte, {len(Bo)} Biología, {len(Co)} Química)")
````

# 📚 Formulación y Resolución del Problema de Exámenes

Esta función `get_solution()` implementa matemáticamente todo nuestro problema de horarios:

1. 🎯 **Crea el modelo** de programación lineal llamado "ExamScheduling"
2. 📉 **Define el objetivo**: minimizar la suma de $Y_{día}$ (usar el menor número de días)
3. 🔒 **Establece las restricciones**:
   | Línea | Restricción | Significado |
   |:-----:|:-----------:|:------------|
   | 1226 | $∑_{m} X_{m,día} ≤ Y_{día}$ | Si hay examen, marca el día como usado |
   | 1227-1230 | $∑_{mod} X_{mod,día} ≤ 1$ | Evita conflictos entre grupos específicos |
   | 1233 | $∑_{día} X_{mod,día} = 1$ | Cada módulo debe programarse exactamente una vez |
4. 🧮 **Resuelve** el problema y retorna la solución óptima

````python
def get_solution(Ac, Ao, Bc, Bo, Cc, Co, times):
    """🎓 Returns the binary variables corresponding to the solution
    of given timetabling problem.
    
    Esta función construye y resuelve el modelo de programación lineal entera
    para el problema de horarios de exámenes universitarios.
    
    Args:
        Ac: The set of core art modules
            # 🎨 Módulos obligatorios de Arte
        Ao: The set of optional art modules
            # 🎭 Módulos opcionales de Arte
        Bc: The set of core biology modules
            # 🧬 Módulos obligatorios de Biología
        Bo: The set of optional biology modules
            # 🔬 Módulos opcionales de Biología
        Cc: The set of core chemistry modules
            # ⚗️ Módulos obligatorios de Química
        Co: The set of optional chemistry modules
            # 🧪 Módulos opcionales de Química
        times: The collection of available time slots.
               # ⏰ Franjas horarias disponibles para programar exámenes
    
    Returns:
        A tuple containing the decision variables x and y.
        # 📊 Retorna las variables con la solución óptima
    """
    
    # 📚 Combinamos todos los módulos en una sola lista
    modules = Ac + Ao + Bc + Bo + Cc + Co
    
    # 🧮 Creamos las variables binarias para nuestro modelo
    x, y = get_variables(modules=modules, times=times)
    
    # 🔧 Creamos el problema de programación lineal - indicamos que queremos minimizar
    prob = pulp.LpProblem("ExamScheduling", pulp.LpMinimize)
    
    # 🎯 Definimos la función objetivo: minimizar el número de días utilizados
    # En programación lineal, esto se expresa como minimizar la suma de variables y[día]
    objective_function = sum([y[day] for day in times])
    prob += objective_function
    
    # 📏 Calculamos un factor de escala para la restricción 8.5
    # M es 1/|M| de la formulación matemática
    M = 1 / len(modules)
    
    # 📝 Añadimos todas las restricciones al modelo
    for day in times:
        # ✅ Restricción 8.5: Si hay al menos un examen en un día, marcamos ese día como usado
        # Esta restricción relaciona las variables x con las variables y
        prob += M * sum(x[m][day] for m in modules) <= y[day]
        
        # ⛔ Restricción 8.7: No más de un módulo de Arte en el mismo día
        # Esta es la restricción de que los módulos de Arte comparten estudiantes
        prob += sum([x[mod][day] for mod in Ac + Ao]) <= 1
        
        # ⛔ Restricción 8.9: No más de un módulo de estos grupos en el mismo día
        # Biología y módulos opcionales de Química comparten estudiantes
        prob += sum([x[mod][day] for mod in Bc + Bo + Co]) <= 1
        
        # ⛔ Restricción 8.8: No más de un módulo de estos grupos en el mismo día
        # Biología y módulos obligatorios de Arte comparten estudiantes
        prob += sum([x[mod][day] for mod in Bc + Bo + Ac]) <= 1
        
        # ⛔ Restricción 8.10: No más de un módulo de estos grupos en el mismo día
        # Módulos opcionales de Biología y Química comparten estudiantes
        prob += sum([x[mod][day] for mod in Cc + Co + Bo]) <= 1
    
    # 📋 Restricción 8.6: Cada módulo debe programarse exactamente una vez
    for mod in modules:
        prob += sum([x[mod][day] for day in times]) == 1
    
    # 🧠 Resolvemos el problema usando el solver de PuLP
    # PuLP seleccionará automáticamente el solver disponible
    prob.solve()
    
    # 🏆 Retornamos la solución óptima (variables x e y con sus valores optimizados)
    return x, y
````

# 🔍 Obtención y Verificación de la Solución

Estas celdas muestran cómo:

1. 🧩 **Obtenemos la solución** llamando a la función `get_solution()` con nuestros grupos de módulos
   
2. 📊 **Verificamos resultados específicos** examinando valores concretos de las variables:
   
   | Consulta | Resultado | Significado |
   |:--------:|:---------:|:------------|
   | `x[2][5].value()` | 0.0 | ❌ El módulo 2 (A3) NO está programado para el día 5 |
   | `x[2][9].value()` | 1.0 | ✅ El módulo 2 (A3) SÍ está programado para el día 9 |

Esto demuestra cómo inspeccionar y verificar la solución óptima que generó el modelo matemático. 🎯

````python
# 🧩 Obtenemos la solución óptima del problema de programación lineal entera
# Llamamos a la función get_solution con todos los conjuntos de módulos definidos anteriormente
x, y = get_solution(
    Ac=Ac, Ao=Ao, Bc=Bc, Bo=Bo, Cc=Cc, Co=Co, times=times
)

# 🔍 Verificamos algunos resultados específicos para entender la solución
# Por ejemplo, comprobamos si el módulo 2 (que es A3, un módulo opcional de arte) está programado en el día 5
print("¿Está el módulo 2 (A3) programado en el día 5?")
print(x[2][5].value())  # 🔢 El método .value() obtiene el valor numérico de la variable en la solución

# 🔍 Verificamos que el módulo 2 está programado en el día 9
print("\n¿Está el módulo 2 (A3) programado en el día 9?")
print(x[2][9].value())  # ✅ Debería ser 1.0 indicando que sí está programado ese día

# 📊 Imprimimos todos los resultados para ver la solución completa
print("\n📋 SOLUCIÓN COMPLETA DEL PROBLEMA DE PROGRAMACIÓN DE EXÁMENES:")
print("=" * 60)

# 🗓️ Primero, veamos qué días se están utilizando
print("\n🗓️ Días utilizados en la solución óptima:")
for t in times:
    if y[t].value() > 0.5:  # ⚠️ Usamos > 0.5 en lugar de == 1 para evitar problemas de precisión numérica
        print(f"  Día {t}: ✅ Utilizado")
    else:
        print(f"  Día {t}: ❌ No utilizado")

# 📚 Ahora, veamos dónde está programado cada módulo
print("\n📚 Programación de cada módulo:")
modules = Ac + Ao + Bc + Bo + Cc + Co
module_types = {
    m: "🎨 Arte obligatorio" if m in Ac else
       "🎭 Arte opcional" if m in Ao else
       "🧬 Biología obligatoria" if m in Bc else
       "🔬 Biología opcional" if m in Bo else
       "⚗️ Química obligatoria" if m in Cc else
       "🧪 Química opcional" if m in Co else
       "❓ Desconocido"
    for m in modules
}

for m in modules:
    for t in times:
        if x[m][t].value() > 0.5:  # ⚠️ Nuevamente usamos > 0.5 por precisión numérica
            print(f"  Módulo {m} ({module_types[m]}): Programado en el día {t}")

# 📈 Finalmente, mostramos un resumen por día de los exámenes programados
print("\n📈 Resumen de exámenes programados por día:")
for t in times:
    if y[t].value() > 0.5:
        exams_on_day = [m for m in modules if x[m][t].value() > 0.5]
        exams_desc = ", ".join([f"Módulo {m}" for m in exams_on_day])
        print(f"  Día {t}: {exams_desc}")

# 🏆 Calculamos y mostramos algunas estadísticas sobre la solución
print("\n🏆 Estadísticas de la solución:")
days_used = sum(1 for t in times if y[t].value() > 0.5)
print(f"  Total de días utilizados: {days_used}")
print(f"  Total de módulos programados: {len(modules)}")
print(f"  Promedio de exámenes por día: {len(modules)/days_used:.2f}")
````

# 📋 Función para Visualizar el Horario Final

Esta celda:

1. 📝 Define la función `get_schedule()` que **convierte la solución matemática en texto legible** para humanos
2. 🔄 Recorre cada día y:
   - ✅ Verifica si el día está siendo utilizado (`y[day].value() == 1`)
   - 🔍 Busca qué módulos están programados ese día (`x[mod][day].value() == 1`)
   - 📊 Construye una cadena de texto con formato:
     ```
     Día 0: 4, 9
     Día 1: 1, 12
     ```

3. 📱 Muestra los resultados llamando a la función y mostrando el texto resultante con `print(schedule)`

Es esencialmente un **formateador de resultados** que hace que la solución sea fácil de leer y entender. 🎯

````python
def get_schedule(x, y, Ac, Ao, Bc, Bo, Cc, Co, times):
    """📋 Returns a human readable schedule corresponding to the
    solution of given timetabling problem."""
    
    # Importamos el módulo datetime primero
    import datetime
    
    # 📚 Combinamos todos los módulos en una lista para iterar sobre ellos
    modules = Ac + Ao + Bc + Bo + Cc + Co
    
    # 🗺️ Creamos mapeos de nombres descriptivos para los módulos
    module_names = {
        0: "A1 - Historia del Arte", 1: "A2 - Técnicas Artísticas",
        2: "A3 - Arte Contemporáneo", 3: "A4 - Escultura", 4: "A5 - Pintura",
        5: "B1 - Biología Celular", 6: "B2 - Genética",
        7: "B3 - Ecología", 8: "B4 - Zoología",
        9: "Q1 - Química Orgánica", 10: "Q2 - Química Inorgánica",
        11: "Q3 - Bioquímica", 12: "Q4 - Química Analítica", 13: "Q5 - Química Física"
    }
    
    # 🏫 Creamos un string para el encabezado del reporte
    schedule = "🏫 PROBLEMA DE HORARIOS DE EXÁMENES UNIVERSITARIOS 🏫\n"
    schedule += "━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n\n"
    schedule += "🎯 SOLUCIÓN OPTIMIZADA DEL PROBLEMA DE PROGRAMACIÓN DE EXÁMENES\n"
    schedule += "━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n"
    
    # 📊 Contadores para las estadísticas
    subjects_by_day = {}
    subjects_by_category = {"Arte": 0, "Biología": 0, "Química": 0}
    obligatory_count = 0
    optional_count = 0
    
    # 📜 Resumen del problema
    schedule += "\n📜 DESCRIPCIÓN DEL PROBLEMA:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    schedule += "   🎓 Universidad con 3 materias: Arte (5 exámenes), Biología (4 exámenes) y Química (5 exámenes)\n"
    schedule += "   🎯 Objetivo: Minimizar el número de días necesarios para programar todos los exámenes\n"
    schedule += "   ⚠️ Restricciones: Estudiantes no pueden presentar exámenes con conflictos el mismo día\n\n"
    
    # 📋 Reglas específicas
    schedule += "📋 REGLAS DE CONFLICTOS:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    schedule += "   ⛔ Regla 1: No pueden coincidir dos exámenes de Arte el mismo día\n"
    schedule += "   ⛔ Regla 2: Exámenes obligatorios de Arte no pueden coincidir con exámenes de Biología\n"
    schedule += "   ⛔ Regla 3: Exámenes opcionales de Química no pueden coincidir con exámenes de Biología\n"
    schedule += "   ⛔ Regla 4: Exámenes obligatorios de Química no pueden coincidir con exámenes opcionales de Biología\n\n"
    
    # 📅 Horario detallado
    schedule += "📅 HORARIO OPTIMIZADO DE EXÁMENES:\n"
    schedule += "━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n"
    
    # 🔄 Iteramos sobre cada día posible
    for day in times:
        # ✅ Verificamos si este día está siendo utilizado en la solución final
        if y[day].value() == 1:
            # Inicializamos contador de exámenes por día
            subjects_by_day[day] = 0
            
            # 📅 Añadimos el encabezado del día al horario con formato mejorado
            schedule += f"\n📆 DÍA {day+1} {'✦' * (day % 3 + 1)} \n"
            schedule += f"┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
            
            day_subjects = []
            # 🧮 Buscamos los módulos programados para este día
            for mod in modules:
                # ✓ Verificamos si el módulo está programado en este día
                if x[mod][day].value() == 1:
                    # Determinamos el tipo de módulo para añadir el ícono correspondiente
                    if mod in Ac:
                        icon = "🎨"  # Arte obligatorio
                        subject_type = "Arte"
                        is_obligatory = True
                    elif mod in Ao:
                        icon = "🎭"  # Arte opcional
                        subject_type = "Arte"
                        is_obligatory = False
                    elif mod in Bc:
                        icon = "🧬"  # Biología obligatoria
                        subject_type = "Biología"
                        is_obligatory = True
                    elif mod in Bo:
                        icon = "🔬"  # Biología opcional
                        subject_type = "Biología"
                        is_obligatory = False
                    elif mod in Cc:
                        icon = "⚗️"  # Química obligatoria
                        subject_type = "Química"
                        is_obligatory = True
                    elif mod in Co:
                        icon = "🧪"  # Química opcional
                        subject_type = "Química"
                        is_obligatory = False
                    else:
                        icon = "📚"  # Genérico para cualquier otro módulo
                        subject_type = "Otro"
                        is_obligatory = False
                    
                    # Actualizamos contadores
                    subjects_by_day[day] += 1
                    subjects_by_category[subject_type] += 1
                    if is_obligatory:
                        obligatory_count += 1
                    else:
                        optional_count += 1
                    
                    # Añadimos el nombre completo del módulo
                    subject_name = module_names.get(mod, f"Módulo {mod}")
                    status_icon = "⭐" if is_obligatory else "☆"
                    time_icon = "🕓 MAÑANA" if subjects_by_day[day] % 2 == 0 else "🕘 TARDE"
                    
                    day_subjects.append(f"   {icon} {status_icon} {time_icon} | {subject_name}")
            
            # Añadimos los módulos ordenados
            for subject in day_subjects:
                schedule += f"{subject}\n"
            
            if not day_subjects:
                schedule += "   📭 No hay exámenes programados para este día\n"
    
    # 🔄 Añadimos un resumen estadístico detallado
    days_used = sum(1 for t in times if y[t].value() == 1)
    total_modules = len(modules)
    
    schedule += "\n━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n"
    schedule += "📊 ANÁLISIS DE LA SOLUCIÓN 📊\n"
    schedule += "━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n\n"
    
    # 🏆 Resultados principales
    schedule += "🏆 RESULTADOS PRINCIPALES:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    optimization = ((len(times) - days_used) / len(times)) * 100
    schedule += f"   ✅ Se han programado exitosamente {total_modules} exámenes\n"
    schedule += f"   ⏱️ Se utilizaron {days_used} días de los {len(times)} posibles ({optimization:.1f}% de optimización)\n"
    schedule += f"   📊 Promedio: {total_modules / days_used:.2f} exámenes por día\n\n"
    
    # Distribución por categoría
    schedule += "🔍 DISTRIBUCIÓN POR MATERIA:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    for category, count in subjects_by_category.items():
        if category == "Arte":
            cat_icon = "🎨"
            details = f"(2 obligatorios, 3 opcionales)"
        elif category == "Biología":
            cat_icon = "🧬"
            details = f"(2 obligatorios, 2 opcionales)"
        elif category == "Química":
            cat_icon = "⚗️"
            details = f"(2 obligatorios, 3 opcionales)"
        else:
            cat_icon = "📚"
            details = ""
        percentage = (count / total_modules) * 100
        schedule += f"   {cat_icon} {category}: {count} exámenes {details} ({percentage:.1f}%)\n"
    
    # Distribución por tipo (obligatorio/opcional)
    schedule += "\n📌 DISTRIBUCIÓN POR TIPO:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    ob_percentage = (obligatory_count / total_modules) * 100
    op_percentage = (optional_count / total_modules) * 100
    schedule += f"   ⭐ Obligatorios: {obligatory_count} exámenes ({ob_percentage:.1f}%)\n"
    schedule += f"   ☆ Opcionales: {optional_count} exámenes ({op_percentage:.1f}%)\n"
    
    # Distribución por día
    if subjects_by_day:
        schedule += "\n📅 CARGA DE EXÁMENES POR DÍA:\n"
        schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
        max_count = max(subjects_by_day.values()) if subjects_by_day else 0
        for day, count in sorted(subjects_by_day.items()):
            bar = "█" * count + "░" * (max_count - count)
            percentage = (count / total_modules) * 100
            schedule += f"   Día {day+1}: {bar} {count} exámenes ({percentage:.1f}%)\n"
    
    # Visualización de conflictos evitados
    schedule += "\n🚫 CONFLICTOS EVITADOS:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    schedule += "   ✓ Ningún examen de Arte coincide con otro examen de Arte\n"
    schedule += "   ✓ Ningún examen obligatorio de Arte coincide con exámenes de Biología\n"
    schedule += "   ✓ Ningún examen opcional de Química coincide con exámenes de Biología\n"
    schedule += "   ✓ Ningún examen obligatorio de Química coincide con exámenes opcionales de Biología\n"
    
    # 🎯 Añadimos conclusión
    schedule += "\n🎓 CONCLUSIÓN:\n"
    schedule += "┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈┈\n"
    schedule += f"   ✨ ¡Se ha encontrado una solución óptima que minimiza el número de días necesarios!\n"
    schedule += f"   📚 Todos los {total_modules} exámenes han sido programados exitosamente\n"
    schedule += f"   ⏱️ Se ha logrado reducir de {len(times)} a {days_used} días la programación\n"
    schedule += f"   🧩 La solución respeta todas las restricciones de conflictos entre exámenes\n"
    
    # Añadimos la fecha del reporte correctamente
    current_time = datetime.datetime.now().strftime('%d/%m/%Y %H:%M')
    schedule += f"\n⏱️ Reporte generado: {current_time}\n"
    schedule += "━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━\n"
    
    # 📊 Retornamos el horario completo formateado
    return schedule

# 📲 Obtenemos y mostramos el horario formateado
schedule = get_schedule(
    x=x,
    y=y,
    times=times,
    Ac=Ac,
    Ao=Ao,
    Bc=Bc,
    Bo=Bo,
    Cc=Cc,
    Co=Co,
)

# 🖨️ Imprimimos el horario para visualizarlo
print(schedule)
````
