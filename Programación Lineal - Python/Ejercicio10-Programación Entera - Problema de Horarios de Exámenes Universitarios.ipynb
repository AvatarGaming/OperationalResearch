## ğŸ“ Las reglas que debemos seguir

## ğŸ§® Â¿CÃ³mo resolvemos este problema?

Para resolver este problema usamos algo llamado "variables de decisiÃ³n":

- Creamos una variable X para cada combinaciÃ³n de examen y dÃ­a
- Por ejemplo, $X_{(A1,DÃ­a1)} = 1$ significa "El examen A1 se programa en el DÃ­a 1"
- $X_{(A1,DÃ­a1)} = 0$ significa "El examen A1 NO se programa en el DÃ­a 1"

TambiÃ©n creamos otra variable Y para cada dÃ­a:
- $Y_{(DÃ­a1)} = 1$ significa "Usamos el DÃ­a 1 para algÃºn examen"
- $Y_{(DÃ­a1)} = 0$ significa "No usamos el DÃ­a 1 para ningÃºn examen"

## ğŸ¯ Nuestro objetivo

Queremos minimizar el nÃºmero total de dÃ­as usados. MatemÃ¡ticamente, esto es:

$$\text{Minimizar: } \sum_{t=1}^{14} Y_t = Y_1 + Y_2 + Y_3 + ... + Y_{14}$$

## ğŸš§ Restricciones en palabras simples

1. **Cada examen debe programarse exactamente una vez** âœ…  
   Si tenemos el examen A1, debe aparecer en exactamente un dÃ­a, ni mÃ¡s ni menos.
   
   $$\sum_{t=1}^{14} X_{(A1,t)} = 1$$

2. **Si usamos un dÃ­a, debemos marcar ese dÃ­a como "usado"** ğŸ”„  
   Si programamos cualquier examen en el DÃ­a 3, entonces $Y_{(DÃ­a3)}$ debe ser 1.
   
   $$\text{Si } X_{(cualquier_examen,DÃ­a3)} = 1, \text{ entonces } Y_{(DÃ­a3)} = 1$$

3. **No podemos poner exÃ¡menes con conflictos en el mismo dÃ­a** âŒ  
   Por ejemplo, no podemos poner A1 y A2 en el mismo dÃ­a (por la Regla 1).
   
   $$X_{(A1,DÃ­a1)} + X_{(A2,DÃ­a1)} \leq 1 \text{ para cualquier dÃ­a}$$

## ğŸ“Š Â¿QuÃ© es la programaciÃ³n lineal entera?

Es una tÃ©cnica matemÃ¡tica para encontrar la mejor soluciÃ³n posible a problemas como este. En lugar de probar todas las combinaciones posibles (que serÃ­an muchÃ­simas), usamos algoritmos especiales que encuentran la soluciÃ³n Ã³ptima rÃ¡pidamente.

### FormulaciÃ³n matemÃ¡tica generalizada:

$$
\begin{align*}
\text{Minimizar: } & \sum_{t=1}^{T} Y_t \\
\text{Sujeto a: } & \sum_{t=1}^{T} X_{m,t} = 1 \quad \forall m \in \text{ExÃ¡menes} \\
& X_{m,t} \leq Y_t \quad \forall m \in \text{ExÃ¡menes}, \forall t \in \text{DÃ­as} \\
& \sum_{m \in \text{Grupo de conflicto}} X_{m,t} \leq 1 \quad \forall t \in \text{DÃ­as} \\
& X_{m,t}, Y_t \in \{0,1\} \quad \forall m \in \text{ExÃ¡menes}, \forall t \in \text{DÃ­as}
\end{align*}
$$

Esta es la tÃ©cnica que utilizan ğŸ« universidades reales, âœˆï¸ aerolÃ­neas, ğŸ¥ hospitales y muchas empresas para resolver problemas complejos de programaciÃ³n y asignaciÃ³n de recursos.

## ğŸ† Beneficios de esta soluciÃ³n

- ğŸ§  Evita conflictos para los estudiantes que toman mÃºltiples asignaturas
- ğŸ“ Asegura que cada examen se programe exactamente una vez
- ğŸ”„ Permite ajustes si cambian las restricciones o se aÃ±aden mÃ¡s exÃ¡menes

# ğŸ“Š ExplicaciÃ³n del CÃ³digo Python para el Problema de ExÃ¡menes

Esta funciÃ³n `get_variables()` crea las variables de decisiÃ³n necesarias para resolver nuestro problema de horarios.

## ğŸ”¢ Variables Binarias

La funciÃ³n genera dos tipos de variables binarias:

- **`x[m,t]`**: Indica si el mÃ³dulo `m` estÃ¡ programado en el tiempo `t` (1 = sÃ­, 0 = no).
- **`y[t]`**: Indica si se usa el tiempo `t` (1 = sÃ­, 0 = no).

## ğŸ§© Uso de la Biblioteca `pulp`

La funciÃ³n utiliza la biblioteca `pulp`, una herramienta poderosa para resolver problemas de programaciÃ³n lineal.

## ğŸ“ Estructura de las Variables

| Variable | Significado               | Valores |
|:--------:|:-------------------------:|:-------:|
| `x[m,t]` | Â¿Examen `m` en tiempo `t`? | 0 Ã³ 1   |
| `y[t]`   | Â¿Se usa el tiempo `t`?     | 0 Ã³ 1   |

Estas variables corresponden exactamente a las $X_{m,t}$ y $Y_t$ de nuestra formulaciÃ³n matemÃ¡tica. âœ…

````python
# ğŸ“š Importamos la biblioteca PuLP para resolver problemas de programaciÃ³n lineal
import pulp

# ğŸ§® DefiniciÃ³n de la funciÃ³n para crear variables de decisiÃ³n
def get_variables(modules, times):
    """ğŸ”¢ Returns the binary variables for a given timetabling
    problem.
    
    Esta funciÃ³n genera las variables de decisiÃ³n binarias necesarias
    para modelar nuestro problema de programaciÃ³n de exÃ¡menes.
    
    Args:
        modules: The complete collection of modules to be
                timetabled. 
                # ğŸ“ Lista de todos los exÃ¡menes que necesitamos programar (A1, A2, B1, etc.)
        
        times: The collection of available time slots.
               # â° Lista de franjas horarias disponibles (dÃ­a1, dÃ­a2, etc.)
    
    Returns:
        A tuple containing the decision variables x and y.
        # ğŸ”„ Devuelve las variables binarias X y Y del modelo
    """
    
    # ğŸ“Š Forma dimensional para la variable X (matriz mÃ³dulos x tiempos)
    # Cada X[m,t] representa si el mÃ³dulo m estÃ¡ programado en el tiempo t
    xshape = (modules, times)
    
    # âœ… Creamos variables X[m,t] de tipo binario (0 o 1)
    # X[m,t] = 1 significa "el mÃ³dulo m estÃ¡ programado en el tiempo t"
    # X[m,t] = 0 significa "el mÃ³dulo m NO estÃ¡ programado en el tiempo t"
    # En programaciÃ³n lineal entera, estas son nuestras variables de decisiÃ³n principales
    x = pulp.LpVariable.dicts("X", xshape, cat=pulp.LpBinary)
    
    # ğŸ—“ï¸ Creamos variables Y[t] de tipo binario (0 o 1)
    # Y[t] = 1 significa "la franja horaria t estÃ¡ siendo utilizada"
    # Y[t] = 0 significa "la franja horaria t NO estÃ¡ siendo utilizada"
    # En nuestro modelo, intentaremos minimizar la suma de estas variables
    y = pulp.LpVariable.dicts("Y", times, cat=pulp.LpBinary)
    
    # ğŸ”„ Devolvemos ambas variables para usarlas en el modelo de optimizaciÃ³n
    return x, y
````

# ğŸ§© InicializaciÃ³n de Datos para el Problema de ExÃ¡menes

Esta celda define los grupos de mÃ³dulos y crea las variables del problema.

## ğŸ¨ Conjuntos de MÃ³dulos

Se definen los siguientes conjuntos de mÃ³dulos:

| Variable | Significado           | Elementos       |
|:--------:|:---------------------:|:---------------:|

| `Bc`     | BiologÃ­a obligatoria  | `[5, 6]`        |

| `Cc`     | QuÃ­mica obligatoria   | `[9, 10]`       |

## ğŸ“‹ Lista Completa de MÃ³dulos

Se crea una lista combinando todos los grupos de mÃ³dulos definidos anteriormente.

Se generan **14 franjas horarias potenciales** utilizando `times = range(14)`.

## ğŸ”¢ CreaciÃ³n de Variables

Se llama a la funciÃ³n `get_variables()` para crear las variables binarias:

````python
# ğŸ¨ Definimos los mÃ³dulos de Arte (A) obligatorios (c = core)
Ac = [0, 1]  # Estos corresponden a los exÃ¡menes A1 y A2 en nuestro problema
# ğŸ­ Definimos los mÃ³dulos de Arte opcionales (o = optional)
Ao = [2, 3, 4]  # Estos corresponden a los exÃ¡menes A3, A4 y A5
# ğŸ§¬ Definimos los mÃ³dulos de BiologÃ­a (B) obligatorios
Bc = [5, 6]  # Estos corresponden a los exÃ¡menes B1 y B2
# ğŸ”¬ Definimos los mÃ³dulos de BiologÃ­a opcionales
Bo = [7, 8]  # Estos corresponden a los exÃ¡menes B3 y B4
# âš—ï¸ Definimos los mÃ³dulos de QuÃ­mica (C) obligatorios
Cc = [9, 10]  # Estos corresponden a los exÃ¡menes Q1 y Q2
# ğŸ§ª Definimos los mÃ³dulos de QuÃ­mica opcionales
Co = [11, 12, 13]  # Estos corresponden a los exÃ¡menes Q3, Q4 y Q5

# Imprimir informaciÃ³n sobre los mÃ³dulos definidos
print("ğŸ“‹ INFORMACIÃ“N DE MÃ“DULOS:")
print(f"ğŸ¨ MÃ³dulos de Arte obligatorios (Ac): {Ac}")
print(f"ğŸ­ MÃ³dulos de Arte opcionales (Ao): {Ao}")
print(f"ğŸ§¬ MÃ³dulos de BiologÃ­a obligatorios (Bc): {Bc}")
print(f"ğŸ”¬ MÃ³dulos de BiologÃ­a opcionales (Bo): {Bo}")
print(f"âš—ï¸ MÃ³dulos de QuÃ­mica obligatorios (Cc): {Cc}")
print(f"ğŸ§ª MÃ³dulos de QuÃ­mica opcionales (Co): {Co}")

# ğŸ“š Creamos la lista completa de todos los mÃ³dulos 
# En Python, el operador + con listas concatena los elementos
modules = Ac + Ao + Bc + Bo + Cc + Co  # Lista completa de los 14 mÃ³dulos
print(f"\nğŸ“š Lista completa de mÃ³dulos: {modules}")
print(f"   Total de mÃ³dulos: {len(modules)}")

# â° Creamos las posibles franjas horarias (en el peor caso, necesitarÃ­amos 14 dÃ­as)
# La funciÃ³n range(14) crea una secuencia de nÃºmeros del 0 al 13
times = range(14)  # Representa los posibles dÃ­as para programar exÃ¡menes
print(f"\nâ° Franjas horarias disponibles: {list(times)}")
print(f"   Total de franjas horarias: {len(list(times))}")

# ğŸ§® Generamos las variables binarias para nuestro modelo de programaciÃ³n lineal entera
# x[m,t] = 1 si el mÃ³dulo m estÃ¡ programado en el tiempo t, 0 en caso contrario
# y[t] = 1 si se usa la franja t, 0 en caso contrario
print("\nğŸ§® Generando variables de decisiÃ³n...")
x, y = get_variables(modules=modules, times=times)  # Usamos la funciÃ³n que definimos antes

# Imprimir informaciÃ³n sobre las variables generadas
print("\nğŸ“Š RESUMEN DE VARIABLES:")
print(f"   â€¢ Variables X: {len(modules) * len(list(times))} variables binarias")
print(f"   â€¢ Variables Y: {len(list(times))} variables binarias")
print("\nğŸ“ SIGNIFICADO:")
print("   â€¢ X[m,t] = 1 si el mÃ³dulo m estÃ¡ programado en el tiempo t")
print("   â€¢ Y[t] = 1 si se utiliza la franja horaria t")

# Mostrar distribuciÃ³n por categorÃ­a
print("\nğŸ” DISTRIBUCIÃ“N DE MÃ“DULOS:")
print(f"   â€¢ MÃ³dulos obligatorios: {len(Ac) + len(Bc) + len(Cc)} ({len(Ac)} Arte, {len(Bc)} BiologÃ­a, {len(Cc)} QuÃ­mica)")
print(f"   â€¢ MÃ³dulos opcionales: {len(Ao) + len(Bo) + len(Co)} ({len(Ao)} Arte, {len(Bo)} BiologÃ­a, {len(Co)} QuÃ­mica)")
````

# ğŸ“š FormulaciÃ³n y ResoluciÃ³n del Problema de ExÃ¡menes

Esta funciÃ³n `get_solution()` implementa matemÃ¡ticamente todo nuestro problema de horarios:

1. ğŸ¯ **Crea el modelo** de programaciÃ³n lineal llamado "ExamScheduling"
2. ğŸ“‰ **Define el objetivo**: minimizar la suma de $Y_{dÃ­a}$ (usar el menor nÃºmero de dÃ­as)
3. ğŸ”’ **Establece las restricciones**:
   | LÃ­nea | RestricciÃ³n | Significado |
   |:-----:|:-----------:|:------------|
   | 1226 | $âˆ‘_{m} X_{m,dÃ­a} â‰¤ Y_{dÃ­a}$ | Si hay examen, marca el dÃ­a como usado |
   | 1227-1230 | $âˆ‘_{mod} X_{mod,dÃ­a} â‰¤ 1$ | Evita conflictos entre grupos especÃ­ficos |
   | 1233 | $âˆ‘_{dÃ­a} X_{mod,dÃ­a} = 1$ | Cada mÃ³dulo debe programarse exactamente una vez |
4. ğŸ§® **Resuelve** el problema y retorna la soluciÃ³n Ã³ptima

````python
def get_solution(Ac, Ao, Bc, Bo, Cc, Co, times):
    """ğŸ“ Returns the binary variables corresponding to the solution
    of given timetabling problem.
    
    Esta funciÃ³n construye y resuelve el modelo de programaciÃ³n lineal entera
    para el problema de horarios de exÃ¡menes universitarios.
    
    Args:
        Ac: The set of core art modules
            # ğŸ¨ MÃ³dulos obligatorios de Arte
        Ao: The set of optional art modules
            # ğŸ­ MÃ³dulos opcionales de Arte
        Bc: The set of core biology modules
            # ğŸ§¬ MÃ³dulos obligatorios de BiologÃ­a
        Bo: The set of optional biology modules
            # ğŸ”¬ MÃ³dulos opcionales de BiologÃ­a
        Cc: The set of core chemistry modules
            # âš—ï¸ MÃ³dulos obligatorios de QuÃ­mica
        Co: The set of optional chemistry modules
            # ğŸ§ª MÃ³dulos opcionales de QuÃ­mica
        times: The collection of available time slots.
               # â° Franjas horarias disponibles para programar exÃ¡menes
    
    Returns:
        A tuple containing the decision variables x and y.
        # ğŸ“Š Retorna las variables con la soluciÃ³n Ã³ptima
    """
    
    # ğŸ“š Combinamos todos los mÃ³dulos en una sola lista
    modules = Ac + Ao + Bc + Bo + Cc + Co
    
    # ğŸ§® Creamos las variables binarias para nuestro modelo
    x, y = get_variables(modules=modules, times=times)
    
    # ğŸ”§ Creamos el problema de programaciÃ³n lineal - indicamos que queremos minimizar
    prob = pulp.LpProblem("ExamScheduling", pulp.LpMinimize)
    
    # ğŸ¯ Definimos la funciÃ³n objetivo: minimizar el nÃºmero de dÃ­as utilizados
    # En programaciÃ³n lineal, esto se expresa como minimizar la suma de variables y[dÃ­a]
    objective_function = sum([y[day] for day in times])
    prob += objective_function
    
    # ğŸ“ Calculamos un factor de escala para la restricciÃ³n 8.5
    # M es 1/|M| de la formulaciÃ³n matemÃ¡tica
    M = 1 / len(modules)
    
    # ğŸ“ AÃ±adimos todas las restricciones al modelo
    for day in times:
        # âœ… RestricciÃ³n 8.5: Si hay al menos un examen en un dÃ­a, marcamos ese dÃ­a como usado
        # Esta restricciÃ³n relaciona las variables x con las variables y
        prob += M * sum(x[m][day] for m in modules) <= y[day]
        
        # â›” RestricciÃ³n 8.7: No mÃ¡s de un mÃ³dulo de Arte en el mismo dÃ­a
        # Esta es la restricciÃ³n de que los mÃ³dulos de Arte comparten estudiantes
        prob += sum([x[mod][day] for mod in Ac + Ao]) <= 1
        
        # â›” RestricciÃ³n 8.9: No mÃ¡s de un mÃ³dulo de estos grupos en el mismo dÃ­a
        # BiologÃ­a y mÃ³dulos opcionales de QuÃ­mica comparten estudiantes
        prob += sum([x[mod][day] for mod in Bc + Bo + Co]) <= 1
        
        # â›” RestricciÃ³n 8.8: No mÃ¡s de un mÃ³dulo de estos grupos en el mismo dÃ­a
        # BiologÃ­a y mÃ³dulos obligatorios de Arte comparten estudiantes
        prob += sum([x[mod][day] for mod in Bc + Bo + Ac]) <= 1
        
        # â›” RestricciÃ³n 8.10: No mÃ¡s de un mÃ³dulo de estos grupos en el mismo dÃ­a
        # MÃ³dulos opcionales de BiologÃ­a y QuÃ­mica comparten estudiantes
        prob += sum([x[mod][day] for mod in Cc + Co + Bo]) <= 1
    
    # ğŸ“‹ RestricciÃ³n 8.6: Cada mÃ³dulo debe programarse exactamente una vez
    for mod in modules:
        prob += sum([x[mod][day] for day in times]) == 1
    
    # ğŸ§  Resolvemos el problema usando el solver de PuLP
    # PuLP seleccionarÃ¡ automÃ¡ticamente el solver disponible
    prob.solve()
    
    # ğŸ† Retornamos la soluciÃ³n Ã³ptima (variables x e y con sus valores optimizados)
    return x, y
````

Esto demuestra cÃ³mo inspeccionar y verificar la soluciÃ³n Ã³ptima que generÃ³ el modelo matemÃ¡tico. ğŸ¯

````python
# ğŸ§© Obtenemos la soluciÃ³n Ã³ptima del problema de programaciÃ³n lineal entera
# Llamamos a la funciÃ³n get_solution con todos los conjuntos de mÃ³dulos definidos anteriormente
x, y = get_solution(
    Ac=Ac, Ao=Ao, Bc=Bc, Bo=Bo, Cc=Cc, Co=Co, times=times
)

# ğŸ” Verificamos algunos resultados especÃ­ficos para entender la soluciÃ³n
# Por ejemplo, comprobamos si el mÃ³dulo 2 (que es A3, un mÃ³dulo opcional de arte) estÃ¡ programado en el dÃ­a 5
print("Â¿EstÃ¡ el mÃ³dulo 2 (A3) programado en el dÃ­a 5?")
print(x[2][5].value())  # ğŸ”¢ El mÃ©todo .value() obtiene el valor numÃ©rico de la variable en la soluciÃ³n

# ğŸ” Verificamos que el mÃ³dulo 2 estÃ¡ programado en el dÃ­a 9
print("\nÂ¿EstÃ¡ el mÃ³dulo 2 (A3) programado en el dÃ­a 9?")
print(x[2][9].value())  # âœ… DeberÃ­a ser 1.0 indicando que sÃ­ estÃ¡ programado ese dÃ­a

# ğŸ“Š Imprimimos todos los resultados para ver la soluciÃ³n completa
print("\nğŸ“‹ SOLUCIÃ“N COMPLETA DEL PROBLEMA DE PROGRAMACIÃ“N DE EXÃMENES:")
print("=" * 60)

# ğŸ—“ï¸ Primero, veamos quÃ© dÃ­as se estÃ¡n utilizando
print("\nğŸ—“ï¸ DÃ­as utilizados en la soluciÃ³n Ã³ptima:")
for t in times:
    if y[t].value() > 0.5:  # âš ï¸ Usamos > 0.5 en lugar de == 1 para evitar problemas de precisiÃ³n numÃ©rica
        print(f"  DÃ­a {t}: âœ… Utilizado")
    else:
        print(f"  DÃ­a {t}: âŒ No utilizado")

# ğŸ“š Ahora, veamos dÃ³nde estÃ¡ programado cada mÃ³dulo
print("\nğŸ“š ProgramaciÃ³n de cada mÃ³dulo:")
modules = Ac + Ao + Bc + Bo + Cc + Co
module_types = {
    m: "ğŸ¨ Arte obligatorio" if m in Ac else
       "ğŸ­ Arte opcional" if m in Ao else
       "ğŸ§¬ BiologÃ­a obligatoria" if m in Bc else
       "ğŸ”¬ BiologÃ­a opcional" if m in Bo else
       "âš—ï¸ QuÃ­mica obligatoria" if m in Cc else
       "ğŸ§ª QuÃ­mica opcional" if m in Co else
       "â“ Desconocido"
    for m in modules
}

for m in modules:
    for t in times:
        if x[m][t].value() > 0.5:  # âš ï¸ Nuevamente usamos > 0.5 por precisiÃ³n numÃ©rica
            print(f"  MÃ³dulo {m} ({module_types[m]}): Programado en el dÃ­a {t}")

# ğŸ“ˆ Finalmente, mostramos un resumen por dÃ­a de los exÃ¡menes programados
print("\nğŸ“ˆ Resumen de exÃ¡menes programados por dÃ­a:")
for t in times:
    if y[t].value() > 0.5:
        exams_on_day = [m for m in modules if x[m][t].value() > 0.5]
        exams_desc = ", ".join([f"MÃ³dulo {m}" for m in exams_on_day])
        print(f"  DÃ­a {t}: {exams_desc}")

# ğŸ† Calculamos y mostramos algunas estadÃ­sticas sobre la soluciÃ³n
print("\nğŸ† EstadÃ­sticas de la soluciÃ³n:")
days_used = sum(1 for t in times if y[t].value() > 0.5)
print(f"  Total de dÃ­as utilizados: {days_used}")
print(f"  Total de mÃ³dulos programados: {len(modules)}")
print(f"  Promedio de exÃ¡menes por dÃ­a: {len(modules)/days_used:.2f}")
````

````python
def get_schedule(x, y, Ac, Ao, Bc, Bo, Cc, Co, times):
    """ğŸ“‹ Returns a human readable schedule corresponding to the
    solution of given timetabling problem."""
    
    # Importamos el mÃ³dulo datetime primero
    import datetime
    
    # ğŸ“š Combinamos todos los mÃ³dulos en una lista para iterar sobre ellos
    modules = Ac + Ao + Bc + Bo + Cc + Co
    
    # ğŸ—ºï¸ Creamos mapeos de nombres descriptivos para los mÃ³dulos
    module_names = {
        0: "A1 - Historia del Arte", 1: "A2 - TÃ©cnicas ArtÃ­sticas",
        2: "A3 - Arte ContemporÃ¡neo", 3: "A4 - Escultura", 4: "A5 - Pintura",
        5: "B1 - BiologÃ­a Celular", 6: "B2 - GenÃ©tica",
        7: "B3 - EcologÃ­a", 8: "B4 - ZoologÃ­a",
        9: "Q1 - QuÃ­mica OrgÃ¡nica", 10: "Q2 - QuÃ­mica InorgÃ¡nica",
        11: "Q3 - BioquÃ­mica", 12: "Q4 - QuÃ­mica AnalÃ­tica", 13: "Q5 - QuÃ­mica FÃ­sica"
    }
    
    # ğŸ« Creamos un string para el encabezado del reporte
    schedule = "ğŸ« PROBLEMA DE HORARIOS DE EXÃMENES UNIVERSITARIOS ğŸ«\n"
    schedule += "â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n\n"
    schedule += "ğŸ¯ SOLUCIÃ“N OPTIMIZADA DEL PROBLEMA DE PROGRAMACIÃ“N DE EXÃMENES\n"
    schedule += "â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n"
    
    # ğŸ“Š Contadores para las estadÃ­sticas
    subjects_by_day = {}
    subjects_by_category = {"Arte": 0, "BiologÃ­a": 0, "QuÃ­mica": 0}
    obligatory_count = 0
    optional_count = 0
    
    # ğŸ“œ Resumen del problema
    schedule += "\nğŸ“œ DESCRIPCIÃ“N DEL PROBLEMA:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    schedule += "   ğŸ“ Universidad con 3 materias: Arte (5 exÃ¡menes), BiologÃ­a (4 exÃ¡menes) y QuÃ­mica (5 exÃ¡menes)\n"
    schedule += "   ğŸ¯ Objetivo: Minimizar el nÃºmero de dÃ­as necesarios para programar todos los exÃ¡menes\n"
    schedule += "   âš ï¸ Restricciones: Estudiantes no pueden presentar exÃ¡menes con conflictos el mismo dÃ­a\n\n"
    
    # ğŸ“‹ Reglas especÃ­ficas
    schedule += "ğŸ“‹ REGLAS DE CONFLICTOS:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    schedule += "   â›” Regla 1: No pueden coincidir dos exÃ¡menes de Arte el mismo dÃ­a\n"
    schedule += "   â›” Regla 2: ExÃ¡menes obligatorios de Arte no pueden coincidir con exÃ¡menes de BiologÃ­a\n"
    schedule += "   â›” Regla 3: ExÃ¡menes opcionales de QuÃ­mica no pueden coincidir con exÃ¡menes de BiologÃ­a\n"
    schedule += "   â›” Regla 4: ExÃ¡menes obligatorios de QuÃ­mica no pueden coincidir con exÃ¡menes opcionales de BiologÃ­a\n\n"
    
    # ğŸ“… Horario detallado
    schedule += "ğŸ“… HORARIO OPTIMIZADO DE EXÃMENES:\n"
    schedule += "â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n"
    
    # ğŸ”„ Iteramos sobre cada dÃ­a posible
    for day in times:
        # âœ… Verificamos si este dÃ­a estÃ¡ siendo utilizado en la soluciÃ³n final
        if y[day].value() == 1:
            # Inicializamos contador de exÃ¡menes por dÃ­a
            subjects_by_day[day] = 0
            
            # ğŸ“… AÃ±adimos el encabezado del dÃ­a al horario con formato mejorado
            schedule += f"\nğŸ“† DÃA {day+1} {'âœ¦' * (day % 3 + 1)} \n"
            schedule += f"â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
            
            day_subjects = []
            # ğŸ§® Buscamos los mÃ³dulos programados para este dÃ­a
            for mod in modules:
                # âœ“ Verificamos si el mÃ³dulo estÃ¡ programado en este dÃ­a
                if x[mod][day].value() == 1:
                    # Determinamos el tipo de mÃ³dulo para aÃ±adir el Ã­cono correspondiente
                    if mod in Ac:
                        icon = "ğŸ¨"  # Arte obligatorio
                        subject_type = "Arte"
                        is_obligatory = True
                    elif mod in Ao:
                        icon = "ğŸ­"  # Arte opcional
                        subject_type = "Arte"
                        is_obligatory = False
                    elif mod in Bc:
                        icon = "ğŸ§¬"  # BiologÃ­a obligatoria
                        subject_type = "BiologÃ­a"
                        is_obligatory = True
                    elif mod in Bo:
                        icon = "ğŸ”¬"  # BiologÃ­a opcional
                        subject_type = "BiologÃ­a"
                        is_obligatory = False
                    elif mod in Cc:
                        icon = "âš—ï¸"  # QuÃ­mica obligatoria
                        subject_type = "QuÃ­mica"
                        is_obligatory = True
                    elif mod in Co:
                        icon = "ğŸ§ª"  # QuÃ­mica opcional
                        subject_type = "QuÃ­mica"
                        is_obligatory = False
                    else:
                        icon = "ğŸ“š"  # GenÃ©rico para cualquier otro mÃ³dulo
                        subject_type = "Otro"
                        is_obligatory = False
                    
                    # Actualizamos contadores
                    subjects_by_day[day] += 1
                    subjects_by_category[subject_type] += 1
                    if is_obligatory:
                        obligatory_count += 1
                    else:
                        optional_count += 1
                    
                    # AÃ±adimos el nombre completo del mÃ³dulo
                    subject_name = module_names.get(mod, f"MÃ³dulo {mod}")
                    status_icon = "â­" if is_obligatory else "â˜†"
                    time_icon = "ğŸ•“ MAÃ‘ANA" if subjects_by_day[day] % 2 == 0 else "ğŸ•˜ TARDE"
                    
                    day_subjects.append(f"   {icon} {status_icon} {time_icon} | {subject_name}")
            
            # AÃ±adimos los mÃ³dulos ordenados
            for subject in day_subjects:
                schedule += f"{subject}\n"
            
            if not day_subjects:
                schedule += "   ğŸ“­ No hay exÃ¡menes programados para este dÃ­a\n"
    
    # ğŸ”„ AÃ±adimos un resumen estadÃ­stico detallado
    days_used = sum(1 for t in times if y[t].value() == 1)
    total_modules = len(modules)
    
    schedule += "\nâ”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n"
    schedule += "ğŸ“Š ANÃLISIS DE LA SOLUCIÃ“N ğŸ“Š\n"
    schedule += "â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n\n"
    
    # ğŸ† Resultados principales
    schedule += "ğŸ† RESULTADOS PRINCIPALES:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    optimization = ((len(times) - days_used) / len(times)) * 100
    schedule += f"   âœ… Se han programado exitosamente {total_modules} exÃ¡menes\n"
    schedule += f"   â±ï¸ Se utilizaron {days_used} dÃ­as de los {len(times)} posibles ({optimization:.1f}% de optimizaciÃ³n)\n"
    schedule += f"   ğŸ“Š Promedio: {total_modules / days_used:.2f} exÃ¡menes por dÃ­a\n\n"
    
    # DistribuciÃ³n por categorÃ­a
    schedule += "ğŸ” DISTRIBUCIÃ“N POR MATERIA:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    for category, count in subjects_by_category.items():
        if category == "Arte":
            cat_icon = "ğŸ¨"
            details = f"(2 obligatorios, 3 opcionales)"
        elif category == "BiologÃ­a":
            cat_icon = "ğŸ§¬"
            details = f"(2 obligatorios, 2 opcionales)"
        elif category == "QuÃ­mica":
            cat_icon = "âš—ï¸"
            details = f"(2 obligatorios, 3 opcionales)"
        else:
            cat_icon = "ğŸ“š"
            details = ""
        percentage = (count / total_modules) * 100
        schedule += f"   {cat_icon} {category}: {count} exÃ¡menes {details} ({percentage:.1f}%)\n"
    
    # DistribuciÃ³n por tipo (obligatorio/opcional)
    schedule += "\nğŸ“Œ DISTRIBUCIÃ“N POR TIPO:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    ob_percentage = (obligatory_count / total_modules) * 100
    op_percentage = (optional_count / total_modules) * 100
    schedule += f"   â­ Obligatorios: {obligatory_count} exÃ¡menes ({ob_percentage:.1f}%)\n"
    schedule += f"   â˜† Opcionales: {optional_count} exÃ¡menes ({op_percentage:.1f}%)\n"
    
    # DistribuciÃ³n por dÃ­a
    if subjects_by_day:
        schedule += "\nğŸ“… CARGA DE EXÃMENES POR DÃA:\n"
        schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
        max_count = max(subjects_by_day.values()) if subjects_by_day else 0
        for day, count in sorted(subjects_by_day.items()):
            bar = "â–ˆ" * count + "â–‘" * (max_count - count)
            percentage = (count / total_modules) * 100
            schedule += f"   DÃ­a {day+1}: {bar} {count} exÃ¡menes ({percentage:.1f}%)\n"
    
    # VisualizaciÃ³n de conflictos evitados
    schedule += "\nğŸš« CONFLICTOS EVITADOS:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    schedule += "   âœ“ NingÃºn examen de Arte coincide con otro examen de Arte\n"
    schedule += "   âœ“ NingÃºn examen obligatorio de Arte coincide con exÃ¡menes de BiologÃ­a\n"
    schedule += "   âœ“ NingÃºn examen opcional de QuÃ­mica coincide con exÃ¡menes de BiologÃ­a\n"
    schedule += "   âœ“ NingÃºn examen obligatorio de QuÃ­mica coincide con exÃ¡menes opcionales de BiologÃ­a\n"
    
    # ğŸ¯ AÃ±adimos conclusiÃ³n
    schedule += "\nğŸ“ CONCLUSIÃ“N:\n"
    schedule += "â”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆâ”ˆ\n"
    schedule += f"   âœ¨ Â¡Se ha encontrado una soluciÃ³n Ã³ptima que minimiza el nÃºmero de dÃ­as necesarios!\n"
    schedule += f"   ğŸ“š Todos los {total_modules} exÃ¡menes han sido programados exitosamente\n"
    schedule += f"   â±ï¸ Se ha logrado reducir de {len(times)} a {days_used} dÃ­as la programaciÃ³n\n"
    schedule += f"   ğŸ§© La soluciÃ³n respeta todas las restricciones de conflictos entre exÃ¡menes\n"
    
    # AÃ±adimos la fecha del reporte correctamente
    current_time = datetime.datetime.now().strftime('%d/%m/%Y %H:%M')
    schedule += f"\nâ±ï¸ Reporte generado: {current_time}\n"
    schedule += "â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”\n"
    
    # ğŸ“Š Retornamos el horario completo formateado
    return schedule

# ğŸ“² Obtenemos y mostramos el horario formateado
schedule = get_schedule(
    x=x,
    y=y,
    times=times,
    Ac=Ac,
    Ao=Ao,
    Bc=Bc,
    Bo=Bo,
    Cc=Cc,
    Co=Co,
)

# ğŸ–¨ï¸ Imprimimos el horario para visualizarlo
print(schedule)
````
